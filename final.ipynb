```python
from PyQt5 import QtGui
from PyQt5.QtWidgets import QApplication,  QWidget, QTableWidget, QTableWidgetItem, QVBoxLayout
import sys
class Window(QWidget):
    def __init__(self):
        super().__init__()
 
        self.title = "PyQt5 Tables"
        self.top = 100
        self.left = 100
        self.width = 500
        self.height = 400
 
 
        self.InitWindow()
 
 
    def InitWindow(self):
        self.setWindowIcon(QtGui.QIcon("icon.csv"))
        self.setWindowTitle("Jumlah bencana alam selama 2010-2019")
        self.setGeometry(self.top, self.left, self.width, self.height)
 
        self.creatingTables()
 
 
        self.show()
 
    def creatingTables(self):
        self.tableWidget = QTableWidget()
        self.tableWidget.setRowCount(2)
        self.tableWidget.setColumnCount(11)
 
        self.tableWidget.setItem(0,0, QTableWidgetItem("Abrasi"))
        self.tableWidget.setItem(0,1, QTableWidgetItem("Banjir"))
        self.tableWidget.setItem(0, 2 , QTableWidgetItem("Banjir & Tanah Longsor"))
        self.tableWidget.setItem(0, 3 , QTableWidgetItem("Gempabumi"))
        self.tableWidget.setItem(0, 4 , QTableWidgetItem("Gempabumi & Tsunami"))
        self.tableWidget.setItem(0, 5 , QTableWidgetItem("Kebakaran"))
        self.tableWidget.setItem(0, 6 , QTableWidgetItem("Kekeringan"))
        self.tableWidget.setItem(0, 7 , QTableWidgetItem("Letusa Gunung Api"))
        self.tableWidget.setItem(0, 8 , QTableWidgetItem("Putingbeliung"))
        self.tableWidget.setItem(0, 9 , QTableWidgetItem("Tanah Longsor"))
        self.tableWidget.setItem(0, 10 , QTableWidgetItem("Tsunami"))
 
        self.tableWidget.setItem(1,0, QTableWidgetItem("157"))
        self.tableWidget.setItem(1,1, QTableWidgetItem("5567"))
        self.tableWidget.setItem(1,2, QTableWidgetItem("400"))
        self.tableWidget.setItem(1,3, QTableWidgetItem("180"))
        self.tableWidget.setItem(1,4, QTableWidgetItem("10"))
        self.tableWidget.setItem(1,5, QTableWidgetItem("560"))
        self.tableWidget.setItem(1,6, QTableWidgetItem("658"))
        self.tableWidget.setItem(1,7, QTableWidgetItem("59"))
        self.tableWidget.setItem(1,8, QTableWidgetItem("6003"))
        self.tableWidget.setItem(1,9, QTableWidgetItem("4445"))
        self.tableWidget.setItem(1,10, QTableWidgetItem("4"))
        self.tableWidget.setColumnWidth(1, 200)
         
        self.vBoxLayout = QVBoxLayout()
        self.vBoxLayout.addWidget(self.tableWidget)
        self.setLayout(self.vBoxLayout)
App = QApplication(sys.argv)
window = Window()
sys.exit(App.exec())
```

```python
import matplotlib.pyplot as plt
exp_vals = [157, 5567, 400, 180, 10, 560, 658, 59, 6003, 4445, 4]
exp_labels = ["Abrasi", "Banjir", "Banjir & Tanah Longsor", "Gempabumi", "Gempabumi & Tsunami", "Kebakaran", "Kekeringan", "Letusa Gunung Api", "Putingbeliung", "Tanah Longsor", "Tsunami"]
plt.axis("equal")
plt.title("presentase bencana")
plt.pie(exp_vals,labels=exp_labels,radius=9, autopct='%0.1f%%', shadow=True)
plt.legend()
plt.show()
```
